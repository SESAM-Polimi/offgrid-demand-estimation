```python
import pandas as pd
from collections import Counter
```

```python
df = pd.read_csv('data/Database_ready_v2.csv')
```

## Choosing the target column "presence of light bulb"

```python
target_raw = df["Presence_light_bulb"]
```

### Drivers Columns
removing all columns about appliances (presence and count)

```python
all_drivers = [i for i in df.columns if (not i.startswith("Presence_") and not  i.startswith("m_"))]
# df[drivers[4:]].to_csv("./data/drivers.csv")
# remove first columns about questionare 
all_drivers = all_drivers[4:]
```

### Discard repeated columns 

- Urban/rural
- Type_of_business
- Socio_status_HHH
- Male/female_HHH 
- System_management
- House_ownership_rental_free
- Marital_status_HHH
- Climate_zone_lev_1
- Climate_zone_lev_2
- Climate_zone_lev_3
- Tariff_type
- Connection_type
- Respondent_category
- Tariff_payment
- Tariff_payment_frequency
- Tariff_payment_method

```python
duplicated_drivers = [ "Urban/rural",
                       "Socio_status_HHH",
                       "Male/female_HHH",
                       "System_management",
                       "House_ownership_rental_free",
                       "Marital_status_HHH",
                       "Climate_zone_lev_1",
                       "Climate_zone_lev_2",
                       "Climate_zone_lev_3",
                       "Tariff_type",
                       "Connection_type",
                       "Respondent_category",
                       "Tariff_payment",
                       "Tariff_payment_method",
                       "Tariff_payment_frequency",
                       "Type_of_business"
                       ]
drivers = [i for i in all_drivers if i not in duplicated_drivers]
print("Remaining drivers : ", len(drivers))
```

Generate dataframe with the new drivers list and fill null values with zeros

```python
df = df[drivers].fillna(0)
df
```

```python
df["Respondent_category_Household"]
```

## Textual columns

```python
for driver in drivers:
    if pd.api.types.is_string_dtype(df[driver]):
        print(f"The driver: {driver}  contains strings.")
```

Now all drivers columns contain numeric values so we can train our model. We will export the new dataframe to a new file: 
`data/numeric_data.csv`

```python
df.to_csv("data/numeric_data.csv")
```
