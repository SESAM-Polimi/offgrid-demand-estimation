```python
import geopandas as gpd
import rasterio
from rasterio.mask import mask
import numpy as np

def add_climate_majority(gadm_level_path, tiff_src):

    # Load administrative boundaries (e.g., country, county, district)
    gdf_admin = gpd.read_file(gadm_level_path)
    # Load the GeoTIFF file containing climate data
    with rasterio.open(tiff_src) as src:
        # Initialize an empty list to hold majority climate values
        climate_majority = []

        # Loop through each administrative region
        for _, row in gdf_admin.iterrows():
            # Mask the climate raster to the current region's geometry
            geometry = [row['geometry']]

            # Mask raster data using the district's geometry (GeoDataFrame row's geometry)
            climate_data, transform = mask(src, geometry, crop=True)

            # Flatten the climate data (as it comes in 2D array)
            climate_values = climate_data.flatten()

            # Remove any 'nodata' values if present
            climate_values = climate_values[climate_values != src.nodata]

            # Find the majority climate value (most frequent)
            if climate_values.size > 0:
                values, counts = np.unique(climate_values, return_counts=True)
                majority_value = values[np.argmax(counts)]
                climate_majority.append(majority_value)
            else:
                climate_majority.append(None)

        # Assign the majority climate values to the GeoDataFrame
        gdf_admin['Climate_majority'] = climate_majority
        return gdf_admin
```

```python
gdf_admin = '../playground/data/ESMAP/Ethiopia_L/gadm41_ETH_shp/gadm41_ETH_1.shp'
tiff_src = '/home/awad/Downloads/wc2.1_10m_tmin/wc2.1_10m_tmin_01.tif'

df = add_climate_majority(gdf_admin, tiff_src)
```
