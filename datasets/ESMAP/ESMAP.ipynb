```python
import os
import sys
sys.path.append("../../")  # Adds higher directory to python modules path.
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '1'


from utils import common_modifiers
import numpy as np
from core.ODEDataset import ODEDataset
```

# Merge the datasets:
- Kenya
- Zambia
- ESMAP
- Ethiopia
- Liberia

```python
Kenya = ODEDataset("Kenya").from_csv("../../playground/data/ESMAP/kenya/kenya.csv", index_col=0)

Kenya = Kenya.apply(common_modifiers.add_const_driver("Tariff_payment_frequency", np.nan))
Zambia = ODEDataset("Zambia").from_csv("../../playground/data/ESMAP/zambia/Zambia.csv", index_col=0).apply(
    common_modifiers.rename({"HouseholdID": "ID"}))

Nigeria = ODEDataset("Nigeria").from_csv("../../playground/data/ESMAP/nigeria/Nigeria.csv", index_col=0).apply(
    common_modifiers.rename({"HH_ID": "ID"}))

Ethiopia = ODEDataset("Ethiopia").from_csv("../../playground/data/ESMAP/Ethiopia_L/csv_data/Ethiopia.csv", index_col=0)
Liberia = ODEDataset("Liberia").from_csv("../../playground/data/ESMAP/Liberia_L/Liberia.csv", index_col=0)


Kenya = Kenya.apply(common_modifiers.add_const_driver("source","kenya" ))

Zambia = Zambia.apply(common_modifiers.add_const_driver("source","zambia" ))
Nigeria = Nigeria.apply(common_modifiers.add_const_driver("source","nigeria" ))
Ethiopia = Ethiopia.apply(common_modifiers.add_const_driver("source","ethiopia" ))
Liberia = Liberia.apply(common_modifiers.add_const_driver("source","liberia" ))


```

```python
ESMAP = Kenya.concat(Zambia).concat(Nigeria)
ESMAP = ESMAP.concat(Ethiopia).concat(Liberia)
# %% Save the merged dataset
filename = "../../playground/data/ESMAP/ESMAP_with_source.csv"

ESMAP.to_csv(filename)
```
